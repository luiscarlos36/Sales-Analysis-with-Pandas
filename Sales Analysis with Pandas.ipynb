### Task 2: Add Month column

```python
# Add Month column
all_data['Month'] = all_data['Order Date'].str[0:2]
```

### Task 3: Add a Sales column

The month that had the most sales was December. We can infer that this is due to the fact that Christmas and New Year purchases are made in December, in addition to other festive dates.

## Question 2: What city had the highest number of sales?

### Task 4: Add a City columns

We need to add a column for the cities but we need to be careful as there are cities from different states with the same name. We proceed to differentiate these cities by adding the state they belong to.

```python
# Make some functions to add the City column

def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
```

```python
# Visualize the data
all_data
```

```python
# View the total sales for each City
city_results = all_data.groupby('City').sum()
city_results
```

```python
# Graph the data

cities = all_data['City'].unique()
city_sales = city_results['Sales']

plt.bar(cities, city_sales, color='red', edgecolor='black')
plt.xticks(cities, rotation='vertical')
plt.xlabel('City Name')
plt.ylabel('Sales in millions of USD ($)')
plt.show()
```

As we can see, we have plotted the results but the interpretation was wrong. Our data tells us that the city with the most sales was San Francisco, but our graph tells us that it was Austin, which is incorrect. Let's plot again with the correct data.

```python
# Graph the data

cities = [city for city, df in all_data.groupby('City')]
city_sales = city_results['Sales']

plt.bar(cities, city_sales, color='green', edgecolor='black')
plt.xticks(cities, rotation='vertical')
plt.xlabel('City Name')
plt.ylabel('Sales in millions of USD ($)')
plt.show()
```

The city with the highest number of sales was San Francisco while the city with the lowest number of sales was Portland.

## Question 4: What products are most often sold together?

```python
# Rows duplicated
df = all_data[all_data['Order ID'].duplicated(keep=False)]
df
```

```python
# Add Grouped column
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df
```

```python
# Drop rows that are duplicated
df = df[['Order ID', 'Grouped']].drop_duplicates()
df
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary

from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common(10)
```

```python
# Make the results more presentable

for key, value in count.most_common(10):
    print(key, value)
```

```python
# view results in triads

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 3)))
    
count.most_common(10)

for key, value in count.most_common(10):
    print(key, value)
```

```python
# View the results in pairs

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common(10)

for key, value in count.most_common(10):
    print(key, value)
```

The objects that are sold the most together are cell phones with their respective charging cables or with their headphones and it makes sense since these accessories usually complement each other. It would be strange to see a purchase of an LG Washing Machine with Headphones.

## Question 5: What product sold the most? What do you think it sold the most?

```python
# View the data
all_data
```

```python
# View quantity ordered for each product
product_group = all_data.groupby('Product')
product_group.sum()
```

```python
# View quantity ordered for each product
quantity_ordered = product_group.sum()['Quantity Ordered']
quantity_ordered
```

```python
# Graph the results

products = [product for product, df in product_group]

plt.bar(products, quantity_ordered, color='orange', edgecolor='black')
plt.xticks(products, rotation='vertical')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
# View the mean of the sales for each product
prices = all_data.groupby('Product').mean()['Price Each']
prices
```

```python
# Graph the results

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='yellow', edgecolor='black')
ax2.plot(products, prices, color='brown')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered')
ax2.set_ylabel('Price ($)')
ax1.set_xticklabels(products, rotation='vertical')
plt.show()
```

You can see a trend that the best-selling items are also the cheapest. We also see that the least sold are due to the fact that they are objects that are not bought frequently, that is, a dryer or a washing machine are not bought every day. On the other hand, batteries are objects that are bought very frequently.
